```python
# Importing SKlearn magic

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, accuracy_score
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
# Подправим y_train, y_test так, чтобы там были цифры (на всякий случай)

y_train_scaled = []
y_test_scaled = []

for string in y_train:
    if string == 'opel':
        y_train_scaled.append(1)
    elif string == 'saab':
        y_train_scaled.append(2)
    elif string == 'bus':
        y_train_scaled.append(3)
    else:
        y_train_scaled.append(4)
        
for string in y_test:
    if string == 'opel':
        y_test_scaled.append(1)
    elif string == 'saab':
        y_test_scaled.append(2)
    elif string == 'bus':
        y_test_scaled.append(3)
    else:
        y_test_scaled.append(4)
```

#### 2.6. Bagging.
Перейдем к ансамблям. 

Для построения ансамбля воспользуемся Bagging'ом с числом алгоритмов от 2 до 50. В качестве первого семейства базовых алгоримтов будем использовать линейные модели (т.е. логистическую регрессию), в качестве второго - решающие деревья. 

*Пояснение: Будем строить ансамбль только из моделей из одного семейства, т.е. логистическая регрессия не смешивается с решающими деревьями.*

Для этого можно сгенерировать с помощью метода bootstrap 50 подвыборок из `train` выборки (того же размера, что и исходная), обучить логистическую регрессию и решающее дерево с оптимальными параметрами из предыдущего пункта на каждой из подвыборок и затем усреднить предсказания k моделей.

*Hint: в sklearn может найтись подходящая функция, которая облегчит вам реализацию данного пункта.*

* Постройте график качества классификации и f1-score на `train` и `test` датасетах в зависимости от числа алгоритмов, вошедших в ансамбль.

* Проанализируйте график. Какое количество моделей стоит использовать? Как вы думаете, являются ли параметры решающего дерева, подобранные в предыдущем пункте оптимальными в данном случае?

```python
# Importing

from sklearn.ensemble import BaggingClassifier
```

```python
import warnings

warnings.filterwarnings('always')
```

```python
# Bagging for LogRes

f1_score_list = []
accuracy = []
num_est = np.linspace(2, 30, 19)

# We will use only 2-30 nums of LogRes for bagging because of time :(

# Used estimator
lr_best_bagg = LogisticRegression(C=5.0, class_weight=None, dual=False, fit_intercept=True,
          intercept_scaling=1, max_iter=10000, multi_class='multinomial',
          n_jobs=-1, penalty='l1', random_state=None, solver='saga',
          tol=0.0001, verbose=0, warm_start=False)

for x in num_est:
    k = int(x)
    bagg = BaggingClassifier(base_estimator=lr_best_bagg, n_estimators=k, max_samples=1.0)
    bagg.fit(X_train_scaled, y_train_scaled)
    y_pred_bagg = bagg.predict(X_test_scaled)
    f1_score_list.append(f1_score(y_pred_bagg, y_test_scaled, average='macro'))
    accuracy.append(bagg.score(X_test_scaled, y_test_scaled))   
```

```python
# Plotting

f = plt.figure(figsize = (8, 6))
f.suptitle('Bagging for LogRes')

plt.plot(num_est, f1_score_list, label = 'f1 score')
plt.plot(num_est, accuracy, label = 'accuracy')
plt.legend()
#plt.plot(np.linspace(0., 1., 49), np.linspace(0., 1., 49))
plt.show()
```

```python
#Baggin for trees

f1_score_list_tree = []
accuracy_tree = []
num_est = np.linspace(2, 50, 49)

# Used estimator
best_tree = DecisionTreeClassifier(criterion='gini', splitter='best', max_depth = 42)

for x in num_est:
    k = int(x)
    bagg = BaggingClassifier(base_estimator=best_tree, n_estimators=k, max_samples=1.0)
    bagg.fit(X_train_scaled, y_train_scaled)
    y_pred_bagg = bagg.predict(X_test_scaled)
    f1_score_list_tree.append(f1_score(y_pred_bagg, y_test_scaled, average='macro'))
    accuracy_tree.append(bagg.score(X_test_scaled, y_test_scaled)) 
```

```python
# Plotting

f = plt.figure(figsize = (8, 6))
f.suptitle('Bagging for Trees')

plt.plot(num_est, f1_score_list_tree, label = 'f1 score')
plt.plot(num_est, accuracy_tree, label = 'accuracy')
plt.legend()
plt.show()
```

#### Inside:

1) Количество моделей, которое нужно использовать >= 20, после этого и f1_score, и accuracy выходят на некоторый уровень, вокруг которого просто флуктуирует с дальнейшем увеличением количеством алгоритмов.

2) Параметры решающего дерева скорее всего можно было выбрать по-другому. Из-за того, что наши предсказания по всем алгоритмам усредняются, то глубину дерева можно было взять поменьше, чтобы а) сэкономить время б) деревья были бы ещё более независимы.

```python
# LogRes with bagging

for i in range(1, len(data_list)+1):
    X_res =  data_list[0]
    y_res = label_list[0]
    for j in range(1, i):
        X_res = np.concatenate((X_res, data_list[j]))
        y_res = np.concatenate((y_res, label_list[j]))
    bagg = BaggingClassifier(base_estimator=lr_best_bagg, n_estimators=20, max_samples=1.0)
    bagg.fit(X_res, y_res)
    y_pred = bagg.predict(X_test_scaled)
    f1_score_lr_bagg.append(f1_score(y_pred, y_test_scaled, average='macro'))
    acc_lr_bagg.append(bagg.score(X_test_scaled, y_test_scaled))
    acc_tr_lr_bagg.append(bagg.score(X_train_scaled, y_train_scaled))
```

```python
# Decision tree without bagging

best_tree = DecisionTreeClassifier(criterion='gini', splitter='best', max_depth = 42)

for i in range(1, len(data_list)+1):
    X_res =  data_list[0]
    y_res = label_list[0]
    #print(i)
    for j in range(1, i):
        #print(j)
        X_res = np.concatenate((X_res, data_list[j]))
        y_res = np.concatenate((y_res, label_list[j]))
    best_tree.fit(X_res, y_res)
    y_pred = best_tree.predict(X_test_scaled)
    f1_score_tree.append(f1_score(y_pred, y_test_scaled, average='macro'))
    acc_tree.append(best_tree.score(X_test_scaled, y_test_scaled))
    acc_tr_tree.append(best_tree.score(X_train_scaled, y_train_scaled))
```

```python
# Decision tree with bagging

for i in range(1, len(data_list)+1):
    X_res =  data_list[0]
    y_res = label_list[0]
    #print(i)
    for j in range(1, i):
        #print(j)
        X_res = np.concatenate((X_res, data_list[j]))
        y_res = np.concatenate((y_res, label_list[j]))
    bagg = BaggingClassifier(base_estimator=best_tree, n_estimators=35, max_samples=1.0)
    bagg.fit(X_res, y_res)
    y_pred = bagg.predict(X_test_scaled)
    f1_score_tree_bagg.append(f1_score(y_pred, y_test_scaled, average='macro'))
    acc_tree_bagg.append(bagg.score(X_test_scaled, y_test_scaled))
    acc_tr_tree_bagg.append(bagg.score(X_train_scaled, y_train_scaled))
```

```python
# Random Forest

acc_forest = []
f1_score_forest = []
acc_tr_forest = []

forest = RandomForestClassifier(n_estimators=35)

for i in range(1, len(data_list)+1):
    X_res =  data_list[0]
    y_res = label_list[0]
    for j in range(1, i):
        X_res = np.concatenate((X_res, data_list[j]))
        y_res = np.concatenate((y_res, label_list[j]))
    forest.fit(X_res, y_res)
    y_pred = forest.predict(X_test_scaled)
    f1_score_forest.append(f1_score(y_pred, y_test_scaled, average='macro'))
    acc_forest.append(forest.score(X_test_scaled, y_test_scaled))
    acc_tr_forest.append(forest.score(X_train_scaled, y_train_scaled))
```

```python
# Grap for test accuracy

num = np.arange(1, 11)*10

fig = plt.figure(figsize=(10, 8))

plt.plot(num, acc_lr, label = 'LogRes')
plt.plot(num, acc_lr_bagg, label = 'LogRes bagg')
plt.plot(num, acc_tree, label = 'DecTree')
plt.plot(num, acc_tree_bagg, label = 'DecTree bagg')
plt.plot(num, acc_forest, label = 'RandomForest')

plt.title('Accuracy on test-data')
plt.xlabel('Data part, %')
plt.ylabel('Score')
plt.legend()
plt.show()
```

2) Also analyzing this graphic we could conclude, that if we had more data, tree-based, besides simple decision tree, algorithms would improve and maybe would be more accurate on test-data. Simultaneously logistic regression probaly would not improve it's accuracy more, because it have some strong limit for accuracy and f1-score due to complexity of data.
